**1. What is the difference between a function and a method in Python?**

Ans. In Python, functions and methods are both blocks of reusable code that perform a specific task, but they differ mainly in how they are called and where they are defined:

*  Function:
Definition: A function is defined using the def keyword and is not tied to any object.

Usage: Can be called independently.

Example:

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

* Method:
Definition: A method is a function that is associated with an object, typically defined inside a class.

Usage: Called on an object using the dot (.) notation.

Example:

```python
class Person:
    def greet(self, name):
        return f"Hello, {name}!"

p = Person()
print(p.greet("Jack"))
```

**6. Explain the concept of generators in Python and how they are defined**

Ans. A generator is a special type of iterator that is defined using a function but uses yield instead of return.
*  Key Characteristics of Generators:
 * Generate values on the fly.
 * Do not store the entire sequence in memory.
 * Useful for large datasets or infinite sequences.
 * Automatically creates an iterator object.

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

gen = count_up_to(3)

print(next(gen))
print(next(gen))
print(next(gen))
```

**7. What are the advantages of using generators over regular functions?**

Ans.Advantages of Using Generators Over Regular Functions
 1.  Memory Efficiency (Lazy Evaluation)
 * Generators yield one item at a time, so they don’t load the entiredataset into memory.
 * This is perfect for large files, big data, or infinite sequences.
  
2. Faster Startup Time
* Generators start producing values immediately, unlike regular functions that compute all results before returning.
* Ideal for tasks where you don’t need all results at once.
3. Maintains State Automatically
* Generators pause at each yield and resume where they left off.
* No need to manage loop counters or external state manually.
4. Can Model Infinite Sequences
* Generators can represent infinite streams like sensor data, number generators, etc., because they don't precompute results.
5.  Cleaner, More Readable Code
* Generators help you write cleaner code for custom iterators without needing to define classes with __iter__() and __next__().
6.  Efficient in Pipelines
*  Can be chained together (like Unix pipes) for streaming data transformations.
7.  Useful in Asynchronous Programming
* Generators are the foundation for coroutines and async programming using yield, yield from, and await.

**9.Explain the purpose and usage of the `map()` function in Python.**

Ans. The map() function is a built-in Python function used to apply a function to each item in an iterable (like a list, tuple, etc.) and return a new map object (which is an iterator).

* Purpose of map()
 * To transform or modify all elements in an iterable without writing a loop.
 * It is memory-efficient and more concise than using a for loop.

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squares = map(square, numbers)
print(list(squares))
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

Ans. 1. map() Function
* Purpose:
 * Applies a function to each item in an iterable and returns a new iterator of the results.

```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)
```

2. filter() Function
* Purpose:
* Applies a boolean (True/False) function to each item and returns only those for which the function returns True.

```python
nums = [1, 2, 3, 4]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

3. reduce() Function (from functools module)
*  Purpose:

 * Reduces a sequence to a single value by applying a function cumulatively (pairwise).

```python
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)
```

the list.**

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

    nums = [10, 15, 20, 25, 30]
result = sum_of_even_numbers(nums)
print("Sum of even numbers:", result)
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check up to sqrt(n)
        if n % i == 0:
            return False
    return True
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number.")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms  # total number of terms to generate
        self.count = 0              # tracks how many terms generated so far
        self.a, self.b = 0, 1       # starting values for Fibonacci

    def __iter__(self):
        return self  # an iterator must return itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # end iteration when max_terms is reached
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers

for num in fib:
    print(num, end=' ')
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
for power in powers_of_two(5):
    print(power)
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Removes the newline character
    except FileNotFoundError:
        print(f"File not found: {file_path}")
for line in read_file_line_by_line("sample.txt"):
    print(line)
```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
celsius = [0, 10, 20, 30, 37, 100]

# Convert to Fahrenheit using map and lambda
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

# Output the result
print("Celsius:    ", celsius)
print("Fahrenheit: ", fahrenheit)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    # Keep only characters that are NOT vowels
    return ''.join(filter(lambda char: char not in vowels, text))
input_string = "Beautiful day in the neighborhood!"
result = remove_vowels(input_string)
print("Original String:", input_string)
print("Without Vowels: ", result)
```
